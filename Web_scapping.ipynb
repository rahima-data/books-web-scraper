#  Books Web Scraper

A beginner Python web scraping project that scrapes book titles and prices 
from [books.toscrape.com](http://books.toscrape.com)

## Libraries Used
- `requests` - fetches the webpage
- `BeautifulSoup` - parses the HTML

## Output
Saves a `books.csv` file containing book titles and prices

```python
# requests - lets us fetch a webpage
# BeautifulSoup - lets us read and search through the HTML
import requests
from bs4 import BeautifulSoup
```

```python
# The website we are scraping (built for practice)
url = "http://books.toscrape.com"

# Visit the website and store the response
response = requests.get(url)
response.encoding = "utf-8"  # fixes £ symbol displaying correctly

# Parse the HTML so Python can search through it
soup = BeautifulSoup(response.text, "html.parser")

print(f"Status code: {response.status_code} ✅")
```

```python
# Find all book elements on the page
books = soup.find_all("article", class_="product_pod")
print(f"Found {len(books)} books\n")

# Loop through each book and print title and price
for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    print(f"{title} — {price}")
```

```python
import csv

# Open a new CSV file to write our data into
with open("books.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Price"])  # column headers

    # Loop through each book and save title and price
    for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        writer.writerow([title, price])

print("✅ Data saved to books.csv!")
```
